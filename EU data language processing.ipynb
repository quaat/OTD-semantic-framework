```python
import os
from rdflib import Namespace
from rdflib import BNode
from rdflib.namespace import RDF, RDFS, OWL, DC, FOAF, XSD, SKOS,FOAF
from rdflib import plugin, Graph, Literal, URIRef
from uuid import uuid4

from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.decomposition import TruncatedSVD
DCT = Namespace('http://purl.org/dc/terms/')

docs = []
subdir = "eu"
for file in os.listdir(subdir):
    try:
        graph = Graph()
        graph.parse(os.path.join("eu/", file), format='xml')
        for d in [e for e in graph.objects(None, DCT.description) if e.language == 'en']:
            if d.endswith('.'):
                docs.append(d.lower())
    except:
        pass
print ("generated {0} documents".format(len(docs)))
```

```python
import pandas as pd
import numpy as np

stop = set(stopwords.words('english'))
tf = TfidfVectorizer(stop_words=stop, use_idf=True,ngram_range=(1,6))
tfidf_matrix = tf.fit_transform(docs)
feature_names = tf.get_feature_names()

idx = 0
feature_index = tfidf_matrix[idx,:].nonzero()[1]
tfidf_scores = zip(feature_index, [tfidf_matrix[idx, x] for x in feature_index])

pairs = [(w, s) for w, s in [(feature_names[i], s) for (i, s) in tfidf_scores]]
for pair in sorted(pairs, key=lambda v: v[1], reverse=True)[:25]:
    print (pair)
```

```python
lsa = TruncatedSVD(n_components=60, n_iter=100)
lsa.fit(tfidf_matrix)
terms = tf.get_feature_names()
for i, comp in enumerate(lsa.components_):
    termsInComp = zip (terms, comp)
    sortedTerms = sorted(termsInComp, key=lambda x: x[1], reverse=True) [:15]
    print ("Concept {0}".format(i+1))
    for term in sortedTerms:
        print ("\t"+term[0])
    print ("\n")
```
