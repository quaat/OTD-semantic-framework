```python
search_words = {'bicycle':OTD.Bicycle,
                'bike':OTD.Bicycle,
                'accident':OTD.Accident,
                'air quality':OTD.AirQuality,
                #'bus stop':OTD.BusStop,
                'charging station':OTD.ChargingStation,
                'environment':OTD.EnvironmentInformation,
                'law':OTD.Law,
                'map':OTD.Map,
                'parking':OTD.Parking,
                #'realtime':OTD.RealTime,
                'road':OTD.Road,
                'route plan':OTD.RoutePlan,
                'statistics':OTD.Statistics,
                #'timetable':OTD.Timetable,
                'trail':OTD.Trail,
                'train':OTD.Train
               }

for search_thresh in [0.7, 0.75, 0.8]:
    set_map.build_map(search_thresh)
    for concept_thresh in [0.6, 0.70, 0.75, 0.8, 0.99]:
        print ('Ts={}_Tc={}'.format(str(search_thresh), str(concept_thresh)))
        v = []
        for w in search_words:
            v.append(search_analysis(set_map, w, search_words[w], concept_thresh))        
        statistics(v, 'Ts={}_Tc={}'.format(str(search_thresh), str(concept_thresh)))    
```
